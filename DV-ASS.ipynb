```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("used_cars_data.csv")
df
```

```python
plt.bar(df['Fuel_Type'], df['Price'])
 
plt.title("Bar Chart")
 
plt.xlabel('Fuel Type')
plt.ylabel('Price')
 
plt.show()
```

```python
sns.histplot(x='Price', data=df, kde=True)
plt.show()
```

```python
plt.scatter(df['Year'], df['Seats'])
```

```python
print(df.dtypes)
```

```python

sns.boxplot(x='Fuel_Type', y='Price', data=df)
plt.title('Boxplot of Price by Fuel Type')
plt.show()
```

```python
sns.countplot(x='Owner_Type', data=df)
plt.title('Countplot of Owner Type')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[['Kilometers_Driven', 'Year', 'Seats', 'Price']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Correlation between Numerical Variables')
plt.show()
```

```python
fuel_type_counts = df['Fuel_Type'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(fuel_type_counts, labels=fuel_type_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Fuel Type Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
avg_price_by_transmission = df.groupby('Transmission')['Price'].mean().sort_values(ascending=False)
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_price_by_transmission.index, y=avg_price_by_transmission.values, palette='viridis')
plt.title('Average Price by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Average Price')
plt.show()
```
